Nous allons déjà commencer par mettre en place une approche générique simple pour résoudre ce problème du sac à dos.

<strong style="text-decoration:underline;display:block">Explications</strong>

## 1.1 Modélisation heuristique

L'algorithme [_Hill Climbing_](https://fr.wikipedia.org/wiki/M%C3%A9thode_hill-climbing) (qu'on ferait parfois mieux d'appeler _Valley Descending_) est une solution assez simple à implémenter. En quoi consiste cet algorithme&nbsp;?

<blockquote>
<strong style="text-decoration:underline;display:block">Solution</strong>
    <ol>
        <li>On considère une solution initiale.</li>
        <li>Une boucle génère tous les voisins de la solution courante, et choisit la meilleure comme nouvelle solution courante.</li>
        <li>On sort de la boucle lorsqu'elle ne permet plus de trouver une solution meilleure que la solution courante.</li>
    </ol>

Cet algorithme utilise la même notion de voisinage que certaines métaheuristiques, qu'il faudra donc commencer par définir. On peut se demander si c'est une heuristique, ou une métaheuristique. C'est une question de vocabulaire :
<ul><li>D'une part, on peut estimer que non, car on ne retrouve pas l'étape de diversification typique des métaheuristiques, qui permet parfois de s'extraire d'un optimum local. Ici, il s'agit strictement d'un algorithme glouton, qui effectue une optimisation locale, un peu de la même manière qu'une descente de gradient, mais sur une structure discrète.</li>
<li>D'un autre côté, on peut considérer que oui, puisqu'il s'agit d'un algorithme de recherche locale, exploitant la notion de voisinage. C'est donc une méthode générique (d'où le terme <em>méta</em>) et que l'on peut utiliser pour trouver une solution approchée d'un problème difficile (d'où le terme heuristique). À l'extrême, on peut la considérer comme une métaheuristique faisant uniquement de l'intensification, avec une diversification nulle.</li>
</ul>

La littérature a tendance à considérer que la réponse est "non", on s'en tient donc à cette vision.

D'ailleurs, les élèves ont vu très récemment un algorithme d'optimisation qui fonctionne selon un principe très similaire, et qui d'ailleurs génère la solution optimale (c'est-à-dire qu'il traite des problèmes en utilisant une notion de voisinage dans laquelle un optimum local est aussi un optimum global). C'est le Simplexe. Dans le cas des programmes linéaires en nombres réels (mais pas en entiers), on peut voir le Simplexe comme un algorithme qui considère uniquement un sous-ensemble des solutions réalisables, celles qui correspondent à un sommet du polytope. Le voisinage de solution correspond au fait que deux sommets sont liés par une arête sur le polytope. Le Simplexe part d'un sommet quelconque du polytope (souvent l'origine, de variables nulles), et choisit à chaque itération le sommet voisin qui minimise la fonction économique, jusqu'à arriver à l'optimum. Les programmes linéaires continus sont donc une catégorie de problèmes dans lesquels un optimum local est aussi un optimum global.<br>


Un petit détail&nbsp;: en fait, le Simplexe a une complexité au pire exponentielle (mais dans la plupart des cas, on reste polynomial).
</blockquote>


La première (et la principale) étape de la modélisation d'un problème pour une méta-heuristique est donc la définition de la structure de voisinage (du moins dans le cas d'une méthode par trajectoire). Quelle modification simple d'une solution peut-on considérer dans notre cas de problème de sac à dos&nbsp;?

<blockquote>
<strong style="text-decoration:underline;display:block">Solution</strong>
On considère comme principe général une solution voisine de la solution courante comme une solution valide qui lui est égale à une modification atomique près. 

Pour notre cas, deux solutions sont voisines si l'on peut passer de l'une à l'autre en ajoutant ou retirant un objet du sac.

Cela veut dire que le voisinage d'une solution est l'ensemble des solutions voisines de cette solution. En utilisant cette définition du voisinage d'une solution, il va être possible d'implémenter diverses métaheuristiques pour traiter ce problème.
</blockquote>

Reste à traiter la question de la représentation informatique de nos données. Sachant qu'une solution représente, pour chaque objet, le fait qu'on prenne ou non cet objet, comment va-t-on modéliser informatiquement une telle solution&nbsp;?

<blockquote>
<strong style="text-decoration:underline;display:block">Solution</strong>
Attention, on ne se pose pas encore la question des structures Python, juste de leur représentation numérique).

Une solution est un mot de $n$ bits avec $n$ le nombre d'objets disponibles. La valeur du $i^{ème}$ booléen indique si le $i^{ème}$ objet est présent dans la solution.

D'ailleurs, une fois que l'on considère cette représentation, on peut réexprimer la notion de voisinage selon cette représentation&nbsp;:
Les voisins d'une solution sont tous les mots binaires de $n$ bits ayant une <a href="https://fr.wikipedia.org/wiki/Distance_de_Hamming#D%C3%A9finition_et_exemples">distance de Hamming</a> de 1 avec la solution.
</blockquote>

À partir de là, représenter les poids et valeurs de chaque objet est assez naturel. Quelles structures seront adaptées&nbsp;?

<blockquote>
<strong style="text-decoration:underline;display:block">Solution</strong>
Deux tableaux de $n$ valeurs, avec $n$ le nombre d'objets disponibles, un tableau pour les poids, et un pour les valeurs.
</blockquote>

## 1.2 Structures et fonctions de base

<blockquote class="comment">
<strong style="text-decoration:underline;display:block">Explications</strong>
Nous avons déterminé au-dessus les structures de données qu'on va utiliser, nous pouvons maintenant les implémenter, écrire les fonctions de base associées (calcul de la valeur d'une solution, de son poids), puis implémenter une fonction pour générer les voisins d'une solution. Une fois ces outils créés, on pourra implémenter plusieurs métaheuristiques, les tester, les comparer etc.
</blockquote>

Vous connaissez déjà le principe très général des algorithmes qu'on va implémenter. Il s'agit de générer des solutions valides, d'en évaluer la valeur, et de garder la meilleure. Quelles structures de données Python allons-nous utiliser pour privilégier les performances à l'exécution&nbsp;? Proposez une structure pour&nbsp;:
* une instance du problème
* une solution d'une instance

<blockquote>
<strong style="text-decoration:underline;display:block">Solution</strong>
<ul>
    <li><strong>Instance&nbsp;:</strong> Une instance est constituée&nbsp;:
        <ul>
            <li>du nombre d'objets</li>
            <li>de la capacité du sac</li>
            <li>de tous les objets, chacun défini par son poids et sa valeur</li>
        </ul>
        Les deux premiers ne posent pas de problème, ce sont des entiers. Du point de vue de la performance du code, c'est surtout sur les deux derniers qu'il faut choisir la bonne solution. En l'occurrence, les <em>dictionnaires</em> sont appropriés. En effet, lorsqu'on calculera le poids ou la valeur d'une solution, on n'accédera pas à tous les objets, mais uniquement ceux qui sont présents dans la solution. Et là, la capacité des dictionnaires à accéder à un élément dont on connaît la clé en $O(1)$ (c'est-à-dire en temps constant) grâce à une méthode de hachage va être un atout considérable.<br><br></li>
    <li><strong>Solution&nbsp;:</strong> une instance est représentée par un mot binaire de taille le nombre d'objets.<br>
         La structure la plus adapté est un <em>tuple de booléens</em>. Dans notre cas, on va devoir générer de manière intensive beaucoup de solutions (que ce soit l'une après l'autre dans une approche par trajectoire, ou en même temps dans une méthode par population), qui, une fois créées, seront uniquement évaluées (en termes de valeur et de poids), et éventuellement conservées. Les performances des tuples nous intéressent donc beaucoup plus que la souplesse d'utilisation des listes, dont on n'a finalement pas besoin.
    </li>
</ul>

On ne va pas se compliquer la vie, ces structures seront stockées dans des variables globales. C'est une mauvaise habitude pour un projet d'application, mais ici on fait des expérimentations sur des algorithmes, rien n'ira en production.
</blockquote>

Commençons par nous occuper des instances. Si on veut tester nos fonctions, il faut pouvoir générer aléatoirement une instance (étant donnés un poids et une valeur maximale pour tout objet). Pour le nombre d'objets et la capacité du sac, deux entiers feront l'affaire&nbsp;:

```python
nb_objets = 10
capacite = 20
```

Avec la bibliothèque standard [random](https://docs.python.org/fr/3/library/random.html) de Python, notre fonction de génération d'instance aléatoire sera très simple à écrire. Elle pourra renvoyer un tuple de dictionnaires `(poids, valeur)`, qu'elle aura générés avec des Comprehension. 

```python
import random

def random_objets(poids_max, val_max):
    """
    Cette fonction genere des objets de poids et de valeur
    aleatoires (borne par les valeurs passees en parametre).
    
    Renvoie un tuple de 2 dictionnaires (poids,valeur)
    """

    poids_objets =\
        {i: random.randint(1, poids_max) for i in range(nb_objets)} #SOLUTION
    valeur_objets =\
        {i: random.randint(1, val_max)   for i in range(nb_objets)} #SOLUTION
    
    return poids_objets, valeur_objets

random.seed(a=3) # utiliser un seed explicitement permettra de reproduire les conditions initiales
                 # et de comparer le comportement de différents algorithmes sur une même instance

poids_objets, valeur_objets = random_objets(10, 10)
print(str(poids_objets) + "\n" + str(valeur_objets))
```

Maintenant que nous pouvons générer une instance aléatoire, on va pouvoir implémenter et tester les fonctions de base dont toute métaheuristique a besoin. Lesquelles&nbsp;?

<blockquote>
<strong style="text-decoration:underline;display:block">Solution</strong>
On a besoin de deux fonctions&nbsp;:
<ul>
    <li>Une fonction pour calculer le poids de la solution, c'est-à-dire la somme des poids des objets dans le sac , pour que la métaheuristique vérifie si une solution est valide (si son poids respecte la capacité du sac)&nbsp;:<br>
    $\displaystyle\sum_{i=1}^{n}{w_ix_i}$<br>
<br>
avec (notations reprises du Workshop précédent)
    <ul>
        <li>$w_k\in\mathbb{Z}^{⁺}$ le poids de l'objet $k$</li>
        <li>$x_k\in\{0,1\}$ la variable de décision de l'objet $k$</li>
    </ul>
    <br>
    </li>
    <li>Une fonction calculant la valeur d'une solution, pour que la métaheuristique puisse comparer deux solutions (quelle que soit la métaheuristique envisagée). Le principe est le même que pour la fonction précédente, mais en utilisant la valeur de l'objet plutôt que son poids.
    </li>
</ul>
C'est simple à implémenter, surtout avec la fonction [`sum`](https://www.programiz.com/python-programming/methods/built-in/sum). Mais plutôt que d'utiliser littéralement la formule dont on vient de parler, on va utiliser les capacités de Python&nbsp;: on va construire un sous-ensemble des valeurs de `poids_objets`, qui ne contiennent que les valeurs des objets présents dans le sac. Et il y a un moyen de produire ces valeurs, ce sont les Comprehension (que ce soit sur les listes ou les dictionnaires).
</blockquote>

Ce qu'il faudrait donc, c'est arriver à écrire une Comprehension qui renvoie les indices des objets présents dans le sac. Pour ça, on va ajouter une [List Comprehension conditionnelle](https://riptutorial.com/python/example/767/conditional-list-comprehensions), qui va [énumérer](https://www.guru99.com/python-enumerate-function.html) le contenu du sac, et filtrer uniquement les valeurs `True`. Essayons déjà d'écrire cette Comprehension.

<blockquote class="comment">
<strong style="text-decoration:underline;display:block">Explications</strong>
Comme mentionné au-dessus, les étudiants peuvent proposer d'implémenter le calcul du poids ou de la valeur avec une solution du type&nbsp;:
<pre><code class="cm-s-ipython language-python"><span class="cm-builtin">sum</span>(<span class="cm-variable">valeur_objets</span>[<span class="cm-variable">i</span>]<span class="cm-operator">*</span><span class="cm-variable">val</span> <span class="cm-keyword">for</span> <span class="cm-variable">i</span>, <span class="cm-variable">val</span> <span class="cm-keyword">in</span> <span class="cm-builtin">enumerate</span>(<span class="cm-variable">sac</span>))</code></pre>

    
Cette approche est valide, mais sera un peu plus longue à l'exécution (moins que ce qu'on pourrait imaginer, ceci dit), car on remplace le test d'une valeur booléenne par une multiplication, et ce pour une lisibilité perfectible (même si c'est plus proche du programme linéaire qu'ils ont vu lors du Workshop précédent).
</blockquote>

```python
sac = (True, False, True)

# on teste le générateur avec un List Comprehension
objets =[\
     i for i, val in enumerate(sac) if val] #SOLUTION

print(objets)
```

On obtient bien la liste des indices des objets qui nous intéressent. Maintenant, on va modifier ce générateur pour créer la liste des valeurs de poids des objets présents dans le sac.

<blockquote class="comment">
<strong style="text-decoration:underline;display:block">Explications</strong>
Un indice&nbsp;: ce n'est plus <code>i</code> qu'on veut générer dans la liste, mais <code>poids_objets[i]</code>. Et cette liste, on va la passer à la fonction <code class="cm-s-ipython language-python"><span class="cm-builtin">sum</span></code>.
</blockquote>

On va directement utiliser cette méthode pour implémenter fonction de calcul du poids du sac. Et au passage, on implémente aussi le calcul de la valeur du sac, sur le même principe.

```python
def poids_contenu(sac):
    """
    Cette fonction renvoie la somme des poids des objets dans le sac
    """
    # somme des poids des objets présents dans le sac
    return\
        sum(poids_objets[i] for i, val in enumerate(sac) if val) #SOLUTION

def valeur_contenu(sac):
    """
    Cette fonction renvoie la somme des valeurs des objets dans le sac
    """
    # somme des valeurs des objets présents dans le sac
    return\
        sum(valeur_objets[i] for i, val in enumerate(sac) if val) #SOLUTION
```

Tout fonctionne correctement&nbsp;? Pour le vérifier, le plus simple est de générer une solution aléatoire, et de tester nos deux fonctions sur cette solution. 

```python
import random

random.seed(a=3)
nb_objets = 5
poids_objets, valeur_objets = random_objets(10, 10)

test = tuple(random.choice([False, True]) for _ in range(nb_objets))
for o in range(nb_objets):
    print("%d:(%d,%d) "%(o, poids_objets[o], valeur_objets[o]), end='')

print("")
print([i for i, val in enumerate(test) if val]) # affiche les indices des valeurs True de la solution
print("poids=%d, valeur=%d" %(poids_contenu(test), valeur_contenu(test)))
```

Apparemment, ça marche&nbsp;!

<blockquote class="comment">
<strong style="text-decoration:underline;display:block">Explications</strong>
Il existe pour filtrer les indices des objets présents une implémentation moins lisible que la Comprehension, mais nettement plus rapide. Cette implémentation utilise la fonction <a href="https://docs.python.org/fr/3/library/itertools.html#itertools.compress">compress</a> de la bibliothèque <a href="https://docs.python.org/fr/3/library/itertools.html">itertools</a>. Cette fonction n'accepte pas de dictionnaire, il faut lui passer directement les valeurs, par exemple en les extrayant du dictionnaire avec la méthode <code>values</code>&nbsp;:

<pre><code class="cm-s-ipython language-python"><span class="cm-builtin">sum</span>(<span class="cm-variable">itertools</span>.<span class="cm-property">compress</span>(<span class="cm-variable">poids_objets</span>.<span class="cm-property">values</span>(), <span class="cm-variable">test</span>))</code></pre>

On peut aussi directement remplacer les dictionnaires par des tuples pour représenter les valeurs et les poids, mais le gain est négligeable.

On perd ainsi les performances d'accès du dictionnaire, mais comme la bibliothèque est implémentée en C, on y gagne tout de même nettement (le calcul est 2 à 3 fois plus rapide, et sur l'ensemble de l'exécution de l'algorithme, on observe un gain moyen de 30%). Si les étudiants sont à l'aise avec le code, vous pouvez leur suggérer de chercher dans cette bibliothèque une solution alternative et d'en tester les performances.
</blockquote>

# 2 Heuristique simple
Nous pouvons maintenant nous lancer dans l'implémentation d'un _Hill Climbing_. Pour commencer, il faut implémenter la fonction de calcul de voisinage d'une solution dont on a parlé plus haut. 

## 2.1 Génération de voisinage
Nous voulons donc générer l'ensemble de tous les voisins d'une solution (donc d'un tuple de booléens). Pour être sûr de vous, écrivez cet algorithme de manière plus détaillée. Vous pouvez considérer qu'une fonction de clonage d'une solution est à votre disposition.

<blockquote>
<strong style="text-decoration:underline;display:block">Solution</strong>

Il faut implémenter une fonction qui va prendre en paramètre une solution d'entrée, et qui va générer séquentiellement tous les voisins valides de cette solution. Pour cela, il suffit d'itérer sur le nombre d'objets dans l'instance et de construire, à chaque itération $i$, un tuple clone de la solution considérée, dont la présence dans la solution clonée est inversé par rapport à la solution passée en paramètre.

<em>fonction</em> Voisinage(sac)
<div style="border-left: 1px solid black;padding-left:25px;margin:5px;">
    Créer la liste vide des voisins<br>
    <br>
    <em>Pour</em> chaque objet du problème
    <div style="border-left: 1px solid black;padding-left:25px;margin:5px;">
        Cloner le sac<br>
        <em>Si</em> l'objet est dans le sac<br>
        <div style="border-left: 1px solid black;padding-left:25px;margin:5px;">
            Retirer l'objet du sac cloné
        </div>
        <em>Sinon</em>
        <div style="border-left: 1px solid black;padding-left:25px;margin:5px;">
            Ajouter l'objet dans le sac cloné
        </div>
        <br>
        <em>Si</em> le sac cloné est une solution valide
        <div style="border-left: 1px solid black;padding-left:25px;margin:5px;">
            Ajouter le sac cloné dans la liste des voisins
        </div>
    </div>
    <br>
    Renvoyer la liste des voisins
</div>
</blockquote>

Maintenant, reste à implémenter cet algorithme.

<blockquote class="comment">
<strong style="text-decoration:underline;display:block">Explications</strong>
On va implémenter cette fonction de voisinage comme un <em>générateur</em> (ça rendra l'implémentation de l'algorithme plus économe en mémoire).

Par contre, attention, nous avons choisi les tuples pour représenter nos solutions. Or, les tuples sont immuables, ce qui veut dire qu'on ne peut pas, comme on l'a écrit dans l'algorithme, implémenter un générateur qui clone <em>puis</em> modifie la solution courante, mais qu'il faut générer <em>en une seule fois</em> une copie dont un booléen a été inversé. Pour faire ça, on va <a href="https://docs.python.org/3/library/stdtypes.html#common-sequence-operations">concaténer</a> des tuples <a href="https://www.tutorialspoint.com/How-to-index-and-slice-a-tuple-in-Python">slicés</a> (attention aux bornes du slicing, dans la notation Python, certaines sont ouvertes, d'autres fermées).

Et il ne faut renvoyer que les solutions valides&nbsp;!
</blockquote>

<blockquote class="comment">
<strong style="text-decoration:underline;display:block">Explications</strong>
Il existe une autre approche que celle par slicing, qui utilise un Comprehension avec opérateur ternaire, que l'on passe à la fonction <a href="https://www.geeksforgeeks.org/python-tuple-function/"><code class="cm-s-ipython language-python"><span class="cm-builtin">tuple</span></code></a>&nbsp;:<br>
<pre><code class="cm-s-ipython language-python"><span class="cm-variable">voisin</span> <span class="cm-operator">=</span> <span class="cm-builtin">tuple</span>(<span class="cm-variable">v</span> <span class="cm-keyword">if</span> <span class="cm-variable">i</span> <span class="cm-operator">!=</span> <span class="cm-variable">k</span> <span class="cm-keyword">else</span> <span class="cm-keyword">not</span> <span class="cm-variable">v</span> <span class="cm-keyword">for</span> <span class="cm-variable">i</span>, <span class="cm-variable">v</span> <span class="cm-keyword">in</span> <span class="cm-builtin">enumerate</span>(<span class="cm-variable">sac</span>))</code></pre>

À moins que les étudiants soient très familiers des fonctionnalités avancées de Python, il est peu probable qu'ils aient l'idée de cette approche. Mais si c'est le cas, ils peuvent tester les deux méthodes et comparer leurs performances (en utilisant la même méthode que dans le Workshop 1 ou la corbeille de la séquence 0). Ils devraient s'apercevoir que cette implémentation prend à l'exécution quasiment deux fois plus de temps que la version avec slicing, probablement à cause du <code class="cm-s-ipython language-python"><span class="cm-builtin">enumerate</span>(<span class="cm-variable">sac</span>)</code>. Mais c'est toujours une bonne approche de comparer les temps d'exécution de deux implémentations possibles lorsqu'on fait du calcul scientifique en Python. 
</blockquote>

```python
def voisinage(sac):
    """
    Cette fonction est un générateur de tous les voisins valides d'une solution
    """
    for k in range(len(sac)):
        # un indice : on veut les états des k-1 premiers objets et des |sac|-k-1 derniers objets
        # et l'état inverse de l'objet k (attention aux bornes du slicing)
        voisin = sac[:k] + (not(sac[k]),) + sac[k+1:] #SOLUTION
        if (poids_contenu(voisin) <= capacite):       #SOLUTION
            yield(voisin)                             #SOLUTION

# on teste
test = (True, False, True)
print (test, "\nvoisins :")
for voisin in voisinage(test):
    print(voisin)
```

Cela vous parait-il bon&nbsp;? Pourquoi n'y a-t-il pas plus de solutions&nbsp;?

<blockquote>
Parce que les autres combinaisons de booléens correspondent à des solutions non réalisables, car elles dépassent la capacité du sac.
</blockquote>

## 2.2 Hill Climbing


Nous pouvons enfin nous lancer dans l'implémentation de notre heuristique _Hill Climbing_. Si on devait écrire l'algorithme de manière plus détaillée, que pourrait-on proposer&nbsp;?

<blockquote>
<strong style="text-decoration:underline;display:block">Solution</strong>

<em>fonction</em> HillClimbing(sac initial)
<div style="border-left: 1px solid black;padding-left:25px;margin:5px;">
    sac courant ← sac initial<br>
    <em>Faire</em><br>
    <div style="border-left: 1px solid black;padding-left:25px;margin:5px;">
        Générer le voisinage du sac courant<br>
        <br>
        <em>Pour</em> chaque sac dans le voisinage<br>
        <div style="border-left: 1px solid black;padding-left:25px;margin:5px;">
            <em>Si</em> le sac est de plus grande valeur que le sac courant
                <div style="border-left: 1px solid black;padding-left:25px;margin:5px;">
                    Le nouveau sac devient le sac courant
                </div>
        </div>
    </div>
    <em>Tant que</em> on trouve un nouveau sac<br>
    <br>
    Renvoyer le sac courant
</div>

</blockquote>

Quel défaut a cet algorithme&nbsp;? Comment pourrait-on y remédier&nbsp;?

<blockquote>
<strong style="text-decoration:underline;display:block">Solution</strong>

Il ne prend pas en compte l'existence d'un <em>plateau</em>, c'est-à-dire une zone où la solution courante se stabilise sans se dégrader, puis finit par de nouveau s'améliorer.<br>
    
Il faut faire attention à ce problème, car on ne peut pas simplement remplacer la comparaison de valeurs avec un test $\geq$ au lieu de $>$. En effet, dans le cas de plusieurs optimums locaux voisins les uns des autres, l'algorithme ferait indéfiniment des va-et-vient entre ces solutions.<br>

Une solution serait de faire la comparaison avec $\geq$, mais en autorisant uniquement un certain nombre d'itérations sans amélioration de la solution courante.

Et si les étudiants suggèrent de mettre en place une mémoire pour éviter de revenir sur ses pas, cela revient finalement à implémenter une recherche avec liste Tabou, et c'est la suite du workshop. Ce problème n'est donc pas très important, puisqu'on on le résoudra par la suite avec la méthode par liste Tabou.
</blockquote>

Lançons-nous maintenant dans son implémentation&nbsp;!
</blockquote>

```python
def hill_climbing(solution_initiale):
    """
    1. On part d'un élément de notre ensemble de recherche qu'on déclare élément courant
    2. On considere le voisinage de l'élément courant et on choisit le  meilleur d'entre
       eux comme nouvel élément courant
    3. On boucle jusqu'à convergence sur un optimum local
    """

    solution_courante = solution_initiale
    nouveau = True
    nb_iter = 0 # uniquement utilisé pour l'affichage
    
    while (nouveau):
        nb_iter += 1
        
        # on parcourt tous les voisins de la solution courante pour garder la meilleure
        meilleure_solution = solution_courante                #SOLUTION
        valeur_meilleure = valeur_contenu(meilleure_solution) #SOLUTION
                                                              #SOLUTION
        for voisin in voisinage(solution_courante):           #SOLUTION
            valeur_voisin = valeur_contenu(voisin)            #SOLUTION
            if valeur_voisin > valeur_meilleure:              #SOLUTION
                valeur_meilleure = valeur_voisin              #SOLUTION
                meilleure_solution = voisin                   #SOLUTION
                                                              #SOLUTION
        nouveau = (meilleure_solution != solution_courante)
        solution_courante = meilleure_solution
    
    return solution_courante
```

## 2.3 Exécution de l'heuristique
Testons cet algorithme en prenant comme solution initiale un sac vide, et affichons la solution proposée&nbsp;:

```python
random.seed(a=3)
capacite = 20
nb_objets = 100
poids_objets, valeur_objets=random_objets(10, 10)

sac = (False,)*nb_objets # une fonctionnalité utile en Python

print("optimisation locale")
sol = hill_climbing(sac)
print("valeur finale = " + str(valeur_contenu(sol)) + ", capacite=" + str(poids_contenu(sol)) + "/" + str(capacite))
print([i for i, val in enumerate(sol) if val]) # liste des objets dans le sac
```

Trois objets, ce n'est pas beaucoup&nbsp;!! Ceci dit, on peut peut-être faire quelque chose pour améliorer ça. Réfléchissez&nbsp;: cet algorithme est [déterministe](https://fr.wikipedia.org/wiki/Algorithme_d%C3%A9terministe). Lorsqu'on considère un algorithme d'optimisation comme la recherche tabou, qu'est-ce que ça implique en termes de qualité de solution pour une instance donnée&nbsp;?


question
<blockquote class="comment">
<strong style="text-decoration:underline;display:block">Explications</strong>

</blockquote>



<blockquote>
<strong style="text-decoration:underline;display:block">Solution</strong>

Un algorithme déterministe produira toujours la même solution pour une instance donnée _et une solution de départ donnée_ (c'est lié au caractère déterministe de la Machine de Turing sous-jacente). La qualité de la solution qu'il génère dépend uniquement de la solution de départ.
</blockquote>

Ça ne vous donne pas une idée pour améliorer le résultat produit par l'algorithme&nbsp;? Comment pourrait-on implémenter une telle approche&nbsp;?

<blockquote>
<strong style="text-decoration:underline;display:block">Solution</strong>

On exécute l'algorithme plusieurs fois, en partant de plusieurs solutions initiales différentes, et on garde la meilleure solution produite
</blockquote>

C'est simple à mettre en place, surtout en Python. Quelle fonction devez-vous implémenter&nbsp;? Que fera-t-elle exactement&nbsp;? Un indice, les fonctions de génération aléatoire de Python, par exemple [<code class="cm-s-ipython language-python"><span class="cm-builtin">random_choice</span></code></a>](https://docs.python.org/3/library/random.html#random.choice) devraient vous être utiles.

<blockquote>
<strong style="text-decoration:underline;display:block">Solution</strong>

Il suffit d'implémenter une fonction de génération de solution <strong>valide</strong> aléatoire. Pour cela, on peut générer une suite aléatoire de booléens de taille $n$ (le nombre d'objets disponibles). Cette solution étant potentiellement non réalisable (trop d'objets dans le sac), on va ensuite retirer aléatoirement des objets jusqu'à respecter la contrainte de contenance du sac.
    
Dans l'idéal, un tel générateur produit des solutions selon une distribution uniforme pour que l'étape de diversification fonctionne bien. Ici, la proposition n'est clairement pas uniforme, les solutions ayant beaucoup d'objets seront sur-représentées. Mais générer de manière uniforme une solution valide pour le problème du sac à dos est une problème complexe, qui s'appuie sur des notions de probabilités avancées (chaînes de Markov, méthode de Monte-Carlo…), qui sont hors de portée des étudiants.

Note&nbsp;: De la même manière que pour la génération du voisinage, il est possible de retirer l'objet sélectionné par Comprehension plutôt qu'en utilisant le slicing, mais là aussi ce sera plus lent.
</blockquote>

Allons-y&nbsp;!

<blockquote class="comment">
<strong style="text-decoration:underline;display:block">Explications</strong>
Cette partie permet d'aborder le multi-start qui sera ensuite utilisé sur la métaheuristique. Si les étudiants sont en retard, vous pouvez sauter cette partie pour psser directement à l'implémentation de la métaheuristique. En effet, le Workshop suivant n'utilisera pas le multi-start.
</blockquote>

```python
def random_solution():
    sac = tuple(\
        random.choice([False, True]) for _ in range(nb_objets)) #SOLUTION

    while (poids_contenu(sac) > capacite):
        # liste des indices de valeur `True` dans le sac
        objets_presents = tuple(
            i for i, val in enumerate(sac) if val) #SOLUTION
        
        # choix aléatoire d'un objet à supprimer parmi ceux présents dans le sac
        objet_supprime =\
            random.choice(objets_presents) #SOLUTION
        
        # suppression de l'objet sélectionné aléatoirement
        sac =\
            sac[:objet_supprime] + (False,) + sac[objet_supprime+1:] #SOLUTION
        # l'implémentation :                                                              #SOLUTION
        # sac = tuple(val if i != objet_supprime else False for i, val in enumerate(sac)) #SOLUTION
        # fonctionne aussi mais est un peu plus lente                                     #SOLUTION

    return sac
```

Il ne nous reste plus qu'à tester tout ça. Pour commencer, essayons avec 10 itérations. À chaque itération, affichez le résultat de la recherche (comme l'affichage final l.14 du code à compléter, mais sans afficher le détail de la solution), qu'on puisse voir un peu ce qui se passe.

```python
val_max = 0
for _ in range(5):
    sac = random_solution()                     #SOLUTION
    sol_courante = hill_climbing(sac)           #SOLUTION
    val_courante = valeur_contenu(sol_courante) #SOLUTION
    print("valeur = %d, capacite=%d/%d"         #SOLUTION
          %(valeur_contenu(sol_courante),       #SOLUTION
            poids_contenu(sol_courante),        #SOLUTION
            capacite))                          #SOLUTION
    if (val_courante > val_max):                #SOLUTION
        val_max = val_courante                  #SOLUTION
        sol_max = sol_courante                  #SOLUTION

print("\nvaleur finale = %d, capacite=%d/%d\n" % (val_max, poids_contenu(sol_max),capacite))
print([i for i, val in enumerate(sol_max) if val])
```

On a nettement amélioré le résultat&nbsp;! Et on constate effectivement que l'optimum local trouvé dépend de la solution de départ. En testant plusieurs solutions initiales, on améliore dans certains cas la solution trouvée. Comment s'appelle cette technique&nbsp;?

<blockquote>
<strong style="text-decoration:underline;display:block">Solution</strong>

Le principe de cette approche s'appelle <strong>multi-start</strong>. Contrairement au Hill-Climbing simple, un Hill-Climbing avec multi-start peut être considéré comme une métaheuristique, puisqu'on a bien une étape d'intensification et une étape de diversification. L'intensification correspond à l'étape d'optimisation locale de la solution courante et la diversification au redémarrage d'une nouvelle solution de départ.
</blockquote>

<blockquote>
<strong style="text-decoration:underline;display:block">Explications</strong>
Une autre métaheuristique existante exploite cette idée de tester plusieurs solutions de départ, mais de manière plus sophistiquée. Il s'agit de la méthode GRASP (Greedy Randomized Adaptive Search Procedure), dont le principe général est le suivant&nbsp;:<br />

<em>fonction</em> GRASP()
<div style="border-left: 1px solid black;padding-left:25px;margin:5px;">
<em>Répéter</em>
    <div style="border-left: 1px solid black;padding-left:25px;margin:5px;">
        Générer une solution valide $X$ par glouton randomisé<br />
        Effectuer une recherche locale en partant de $X$<br />
        <em>Si</em> Valeur(Recherche($X$)) > Valeur($meilleur$)
        <div style="border-left: 1px solid black;padding-left:25px;margin:5px;">
            $meilleur \leftarrow X$
        </div>
    </div>
    <em>Jusqu'à</em> la condition d'arrêt<br />
    Renvoyer $meilleur$
</div>
    
La différence entre un simple multi-start et un GRASP est que la méthode GRASP utilise pour construire une solution aléatoire une approche plus élaborée. L'idée générale est de construire de manière itérative la solution de départ, en lui ajoutant aléatoirement un élément parmi les $r$ meilleurs selon un critère de qualité heuristique (insistons sur le fait que ce critère ne peut qu'être heuristique, le problème étant $\mathcal{NP}$-Complet). Cette liste des meilleurs candidats, appelée RCL (Restricted Candidate List), est regénérée à chaque itération&nbsp;:<br />

<em>fonction</em> CréerRCL()
<div style="border-left: 1px solid black;padding-left:25px;margin:5px;">  
    $Solution \leftarrow\{\emptyset\}$<br />
    <em>Répeter</em>
    <div style="border-left: 1px solid black;padding-left:25px;margin:5px;">
        $RCL \leftarrow \{Candidats\}$<br />
        Trier $RCL$ par ordre décroissant de qualité heuristique<br />
        $Candidat\leftarrow$ élément aléatoire dans $RCL[0..r]$<br />
        Ajouter $Candidat$ à $Solution$<br />
    </div>
    <em>Jusqu'à</em> ce que $Solution$ soit complète<br />
    Renvoyer $Solution$
</div>

Ici, on pourrait considérer comme critère de qualité heuristique le rapport valeur/poids des objets. Le choix de ce critère de classement est l'un des critères impactant la performance d'un GRASP. L'autre paramètre est la valeur de $r$. Dans les cas extrêmes, $r=1$ donne une solution heuristique totalement déterministe, $r=n$ (avec $n$ le nombre d'objets) donne une solution totalement aléatoire, ce qui revient à un simple multi-start.
</blockquote>

## 3.1 Recherche avec liste tabou
Une métaheuristique un peu plus maligne qu'un Hill Climbing devrait donner de meilleurs résultats. Parmi toutes les métaheuristiques de la littérature, la méthode par [recherche avec liste tabou](https://fr.wikipedia.org/wiki/Recherche_tabou) (qu'on appelle souvent _méthode tabou_ par raccourcis) est celle qui pourrait être adaptée le plus facilement en partant du code du Hill Climbing. Pourquoi&nbsp;?

<blockquote>
<strong style="text-decoration:underline;display:block">Solution</strong>

La structure générale de l'algorithme est très proche. Il faut juste adapter la recherche dans le voisinage pour inclure la gestion de la liste tabou, et modifier légèrement la condition d'arrêt.
</blockquote>

De quelle structure de données algorithmique avons-nous besoin pour gérer ça&nbsp;?

<blockquote>
<strong style="text-decoration:underline;display:block">Solution</strong>

Une structure FIFO (First in, First Out), c'est-à-dire une file d'attente, avec une taille maximale.
</blockquote>

Pour implémenter ça en Python, quelle structure vous semble adaptée du point de vue des performances&nbsp;?

<blockquote>
<strong style="text-decoration:underline;display:block">Solution</strong>

La structure <code>deque</code> est exactement ce dont on a besoin. Elle effectue les opérations d'ajout en début et en fin en $O(1)$ au lieu du $O(n)$ des autres structures qui pourraient être utilisées (typiquement les listes), ce qui représente un gain non négligeable.

Si les étudiants sont à l'aise avec la programmation, il existe une solution plus efficace à l'exécution sur les listes de grande taille, mais un peu plus subtile à implémenter. L'ajout dans une <code>deque</code> est effectivement très rapide, mais le défaut de la <code>deque</code> est que la recherche d'une valeur se fait en $O(n)$. Pour les grandes tailles de liste (ce vers quoi on va tendre), ce n'est pas idéal, mieux vaut être rapide sur les recherches que sur l'ajout.

Les deux structures qui implémentent une recherche en $O(1)$ sont le <code>set</code> et le <code>dict</code>. Le premier semble idéal, mais il a un gros défaut&nbsp;: l'ordre des éléments n'est pas conservé, ce qui le disqualifie pour implémenter une liste tabou. C'est par contre le cas pour les dictionnaires depuis la version 3.6 de Python. Une solution est donc de créer un dictionnaire dont les clés sont les solutions testées, et avec des valeurs <code>None</code>. La seule petite difficulté est qu'il faut gérer la taille maximale de la liste manuellement, en supprimant l'élément le plus ancien (d'où le besoin de conserver l'ordre) si cette taille maximale est dépassée. Avec une liste tabou de taille 500, le gain varie entre 10% et 30%, selon les instances.

<pre><code class="cm-s-ipython language-python"><span class="cm-comment"># on met à jour la liste tabou</span>
<span class="cm-variable">liste_tabou</span>[<span class="cm-variable">solution_courante</span>] <span class="cm-operator">=</span> <span class="cm-keyword">None</span>
<span class="cm-keyword">if</span> <span class="cm-builtin">len</span>(<span class="cm-variable">liste_tabou</span>)<span class="cm-operator"> &gt; </span><span class="cm-variable">taille_tabou</span>:
    <span class="cm-variable">liste_tabou</span>.<span class="cm-property">pop</span>(<span class="cm-builtin">next</span>(<span class="cm-builtin">iter</span>(<span class="cm-variable">liste_tabou</span>)), <span class="cm-keyword">None</span>)</code></pre>

Une fois de plus, cette technique est un peu délicate à appliquer, à réserver aux personnes à l'aise avec le code.
</blockquote>

Cette fois, on a tout ce qu'il nous faut. Lançons-nous donc dans cette implémentation&nbsp;:

```python
from collections import deque

def recherche_tabou(solution_initiale, taille_tabou, iter_max):
    """
    1. On part d'un élément de notre ensemble de recherche qu'on déclare élément courant
    2. On considère le voisinage de l'élément courant, on choisit le meilleur d'entre
       eux comme nouvel élément courant, parmi ceux absents de la liste tabou, et on l'ajoute
       à la liste tabou
    3. On boucle jusqu'à condition de sortie.
    """
    nb_iter = 0
    liste_tabou =\
        deque((), maxlen = taille_tabou) #SOLUTION

    # variables solutions pour la recherche du voisin optimal non tabou
    solution_courante = solution_initiale
    meilleure = solution_initiale
    meilleure_globale = solution_initiale

    # variables valeurs pour la recherche du voisin optimal non tabou
    valeur_meilleure = valeur_contenu(solution_initiale)
    valeur_meilleure_globale = valeur_meilleure

    
    while (nb_iter < iter_max):
        valeur_meilleure = -1
        
        # on parcourt tous les voisins de la solution courante
        for voisin in voisinage(solution_courante):
            valeur_voisin=valeur_contenu(voisin)                               #SOLUTION
                                                                               #SOLUTION
            # MàJ meilleure solution non taboue trouvée                        #SOLUTION
            if valeur_voisin > valeur_meilleure and voisin not in liste_tabou: #SOLUTION
                valeur_meilleure = valeur_voisin                               #SOLUTION
                meilleure = voisin                                             #SOLUTION

        # on met à jour la meilleure solution rencontrée depuis le debut
        if valeur_meilleure > valeur_meilleure_globale:
            meilleure_globale = meilleure
            valeur_meilleure_globale = valeur_meilleure
            nb_iter = 0
        else:
            nb_iter += 1

        # on passe au meilleur voisin non tabou trouvé
        solution_courante = meilleure

        # on met à jour la liste tabou
        liste_tabou.append(solution_courante)  #SOLUTION

    return meilleure_globale
```

Un test avec une liste tabou de taille 5 sur la même instance que précédemment va nous permettre de comparer le résultat avec celui obtenu par optimisation locale&nbsp;:

```python
nb_objets = 100
capacite = 20
random.seed(a=3)
poids_objets, valeur_objets = random_objets(10, 10)
sac = (False,)*nb_objets
taille_tabou=5
iter_max=30

print("tabou de taille 5")
sol=recherche_tabou(sac, taille_tabou, iter_max)
print("valeur finale = " + str(valeur_contenu(sol)) + ", capacite="+str(poids_contenu(sol)) + "/" + str(capacite))
```

## 3.2 Tabou avec multi-start

D'ailleurs, on peut réutiliser ce principe du multi-start. Implémentez-le en renforçant la diversification (500 itérations, par exemple)&nbsp;:

```python
    random.seed(a=3)
    
    sol_max = None
    val_max = 0                                                     #SOLUTION
    for _ in range (500):                                           #SOLUTION
        sac = random_solution()                                     #SOLUTION
        sol_courante = recherche_tabou(sac, taille_tabou, iter_max) #SOLUTION
        val_courante = valeur_contenu(sol_courante)                 #SOLUTION
        if (val_courante > val_max):                                #SOLUTION
            val_max = val_courante                                  #SOLUTION
            sol_max = sol_courante                                  #SOLUTION
    print("valeur finale = " + str(valeur_contenu(sol_max)))
```

La solution s'en trouve encore améliorée. Ceci dit, le surcoût en temps de calcul est conséquent.

Or, il y a un moyen d'accélérer les calculs sans toucher à grand chose. Certaines des fonctions qu'on a implémentées sont _pures_. Et ça, ça va nous permettre de mettre en place une accélération de l'exécution du code. Avez-vous une idée de comment faire&nbsp;?

<blockquote>
<strong style="text-decoration:underline;display:block">Solution</strong>

 Un indice&nbsp;: il va falloir modifier le compromis temps-mémoire qu'on a fait sur le calcul du voisinage.

Les fonctions de calcul du poids et de la valeur d'une solution sont pures, on peut donc les mémoïser. Pour la fonction de génération du voisinage, c'est aussi faisable (et souhaitable), mais pas en l'état, car cette fonction est un itérable, elle a besoin d'être réinitialisée à chaque nouvel usage. Il faut modifier la fonction pour qu'elle génère une liste qu'elle renverra en une fois. Le code de la métaheuristique n'a par contre pas besoin d'être modifié.<br>

C'est un cas typique où on change de compromis temps-mémoire&nbsp;: on augmente la mémoire utilisée pour réduire le temps de calcul. Ce qui explique pourquoi l'implémentation d'un générateur doit être abandonnée ici, elle correspond à une solution trop économe en mémoire, et pas assez rapide.

Remarque&nbsp;: le fait que ces fonctions soient pures est en fait discutable, car elles dépendent des variables globales qui décrivent l'instance (<code>poids_objets</code>, <code>poids_objets</code>, <code>nb_objets</code>, et <code>capacite</code>). Si les étudiants s'en aperçoivent, on peut faire remarquer que ces valeurs ne changent pas, ce n'est donc pas gênant. Par contre, on va, dans le Workshop suivant, les faire varier pour obtenir des résultats statistiques. À ce moment, ça va poser des problèmes si ces fonctions sont mémoïsées, car elles ne produiront pas les bonnes valeurs. Ce point sera traité le moment venu.
</blockquote>

D'ailleurs, heureusement qu'on a choisi un tuple pour représenter une solution, car on n'aurait pas pu mettre en place cette technique avec des listes (qui ne sont pas hachables).

Avant tout, relançons le calcul en comptant le temps CPU, pour avoir une idée du gain qu'on obtiendra ensuite.

<blockquote class="comment">
<strong style="text-decoration:underline;display:block">Explications</strong>
Une autre possibilité à laquelle les étudiants peuvent penser pour accélérer les calculs consiste à paralléliser le multi-start. Attention, dans ce cas, l'usage des Threads en Python ne sera pas intéressant (en dehors du fait que les générateurs, utilisés pour le voisinage d'une solution, ne sont pas thread-safe)&nbsp;: le <a href="https://wiki.python.org/moin/GlobalInterpreterLock">Global Interpreter Lock</a> empêche l'exécution simultanée de plusieurs Threads (un gros défaut de Python). Seules les E/S peuvent être parallélisées avec les traitements.<br>

Une vraie parallélisation des calculs nécessite de passer par des <a href="https://docs.python.org/3/library/multiprocessing.html">processus lourds</a>, avec toute la problématique que ça implique (dont le ralentissement dû à la commutation de contexte, qui fait qu'une accélération proportionnelle au nombre de cœurs semble hors de portée).<br>

Et dans ce cas, on réduit nettement le bénéfice de la mémoïsation (puisque chaque processus aura son propre cache). Il n'est pas rare qu'un multi-start séquentiel avec mémoïsation soit plus performant qu'un multi-start parallèle par processus lourds avec mémoïsation (ici, c'est le cas).<br>

Il reste une dernière solution, qui consiste à implémenter un partage de cache de mémoïsation entre processus via une mémoire partagée, mais cela commence à devenir assez lourd en termes de développement.
</blockquote>

```python
import time
random.seed(a=3)

sol_max = None
val_max = 0

start = time.process_time()

# multi-start de 500 itérations
for _ in range (500):                                           #SOLUTION
    sac = random_solution()                                     #SOLUTION
    sol_courante = recherche_tabou(sac, taille_tabou, iter_max) #SOLUTION
    val_courante = valeur_contenu(sol_courante)                 #SOLUTION
    if (val_courante > val_max):                                #SOLUTION
        val_max = val_courante                                  #SOLUTION
        sol_max = sol_courante                                  #SOLUTION

stop = time.process_time()

print("valeur finale = " + str(valeur_contenu(sol_max)))
print("calculé en ", stop-start, 's')
```

Testons maintenant cette technique, en redéfinissant les fonctions concernées.

```python
from functools import lru_cache

@lru_cache(maxsize=None)                                             #SOLUTION
def voisinage(sac):                                                  #SOLUTION
    voisins = []                                                     #SOLUTION
    for k in range(len(sac)):                                        #SOLUTION
        voisin = sac[:k] + (not(sac[k]),) + sac[k+1:]                #SOLUTION
        if (poids_contenu(voisin) <= capacite):                      #SOLUTION
            voisins.append(voisin)                                   #SOLUTION
    return voisins                                                   #SOLUTION
                                                                     #SOLUTION
@lru_cache(maxsize=None)                                             #SOLUTION
def poids_contenu(sac):                                              #SOLUTION
    return sum(poids_objets[i] for i, val in enumerate(sac) if val)  #SOLUTION
                                                                     #SOLUTION
@lru_cache(maxsize=None)                                             #SOLUTION
def valeur_contenu(sac):                                             #SOLUTION
    return sum(valeur_objets[i] for i, val in enumerate(sac) if val) #SOLUTION
```

Et testons ça en relançant le calcul précédent&nbsp;:

```python
random.seed(a=3)

sol_max = None
val_max = -1

start = time.process_time()
# multi-start de 500 itérations
for _ in range (500):                                           #SOLUTION
    sac = random_solution()                                     #SOLUTION
    sol_courante = recherche_tabou(sac, taille_tabou, iter_max) #SOLUTION
    val_courante = valeur_contenu(sol_courante)                 #SOLUTION
    if (val_courante > val_max):                                #SOLUTION
        val_max = val_courante                                  #SOLUTION
        sol_max = sol_courante                                  #SOLUTION

stop = time.process_time()

print("valeur finale = " + str(valeur_contenu(sol_max)))
print("calculé en ", stop-start, 's')
```

L'accélération est impressionnante, le temps de calcul a été quasiment divisé par 4&nbsp;! Et plus on fera de calculs sur cette instance, plus on accélérera (en échange d'une consommation mémoire plus importante). Pour vous en convaincre, il suffit de relancer le même calcul, mais avec un seed différent&nbsp;:
